# **Visualization 1 Number of terrorist attacks by year**

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='iyear', data=gtd_data, palette='viridis')
plt.title('Number of Terrorist Attacks by Year')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45)
plt.show()
```

**Conclusion:**
The visualization demonstrates the number of terrorist attacks over the years, providing insights into the trend of terrorist activities globally. Some key observations from the visualization include

*]There has been a notable increase in terrorist attacks over the years, with peaks observed in certain periods.

*]The trend appears to fluctuate, indicating varying levels of terrorist activities across different years.
*]Further analysis is required to understand the underlying factors contributing to the fluctuations in terrorist attacks over time.

# **Visualization 5  Number of attacks over the years for different attack types**

```python
 plt.figure(figsize=(12, 6))
attack_types = gtd_data['attacktype1_txt'].unique()
for attack_type in attack_types:
    attacks_by_year = gtd_data[gtd_data['attacktype1_txt'] == attack_type].groupby('iyear').size()
    sns.lineplot(x=attacks_by_year.index, y=attacks_by_year.values, label=attack_type)

plt.title('Number of Attacks by Year for Different Attack Types')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.legend(title='Attack Type', loc='upper left', bbox_to_anchor=(1,1))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```


**Conclusion:**
The visualization illustrates the number of terrorist attacks over the years for different attack types, providing insights into the temporal trends of various attack methods employed by terrorist groups. Key observations from the visualization include:

*Different attack types exhibit distinct patterns in their frequency over time, reflecting changes in the tactics and strategies of terrorist organizations.

*Certain attack types may show fluctuations or peaks in specific years, indicating periods of heightened activity or shifts in terrorist tactics.

*Understanding the temporal dynamics of attack types can inform counterterrorism efforts and help anticipate future trends in terrorist activities.
